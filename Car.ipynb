```python
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup as bs
```

```python
website = 'https://www.parkers.co.uk/cars-for-sale/used/?_qs=Uk4oUHJpY2UsNTAwMC4wLDEwMDAwMDAwLjAp'
```

```python
response = rq.get(website)
```

```python
response.status_code
```

```python
soup = bs(response.content, 'html.parser')
```

```python
print(soup)
```

```python
cars = soup.find_all(class_="result-item")
cars
```

```python
name = soup.find(class_="panel__primary-link").text.strip()
name
```

```python
year = soup.find(class_="for-sale-result-item__specs").find_all(class_="for-sale-result-item__specs__bullet")[0].text
year
```

```python
mileage = soup.find(class_="for-sale-result-item__specs").find_all(class_="for-sale-result-item__specs__bullet")[1].text
mileage
```

```python
transmission = soup.find(class_="for-sale-result-item__specs").find_all(class_="for-sale-result-item__specs__bullet")[2].text
transmission
```
